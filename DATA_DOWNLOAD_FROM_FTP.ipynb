# this code will download all the xml files from the ftp path

```python
import ftplib
import os

# FTP server details
ftp_server = "10.10.253.10"
ftp_user = "edf.ratan.jha"
ftp_password = "Gfsdser#$431123"
ftp_directory = "/BillingDashboardFiles/"

# Local directory where you want to save the files
local_directory = r"C:\Users\Ratan Kumar Jha\Desktop\BILLING_REQUEST_AND_REPLY\DOWNLOAD FROM FTP"

# Connect to the FTP server
ftp = ftplib.FTP(ftp_server)
ftp.login(user=ftp_user, passwd=ftp_password)

print("FTP login successful")

# Change to the desired directory on the FTP server
ftp.cwd(ftp_directory)

# List files in the directory
files = ftp.nlst()
print("Files in directory:", files)

# Filter files to include only .xml files
xml_files = [file for file in files if file.lower().endswith('.xml')]

if not xml_files:
    print(f"No .xml files found in the directory {ftp_directory}")
else:
    for file_name in xml_files:
        local_path = os.path.join(local_directory, file_name)
        
        # Open the local file for writing in binary mode
        with open(local_path, 'wb') as local_file:
            # Define a callback function to handle the data
            def write_data(data):
                local_file.write(data)
            
            # Retrieve the file from the FTP server
            ftp.retrbinary(f"RETR {file_name}", write_data)

        print(f"File downloaded successfully and saved to {local_path}")

# Close the connection
ftp.quit()
print("FTP connection closed")
```
